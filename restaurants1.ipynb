IMPORTING THE USEFUL LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

IMPORTING THE DATASET

```python
filepath=r'Dataset .csv'
dataset=pd.read_csv(filepath)
```

MAKING A DATAFRAME FOR USEFUL FEATURES

```python
Usefuldataset = dataset[['Restaurant ID','Restaurant Name','Cuisines','Price range','Aggregate rating','Votes']]
```

```python
Usefuldataset.head() 
```

SHOWING AGGREGATE RATING OF ALL THE RESTAURANTS

```python
plt.figure(figsize=(10, 6))
sns.histplot(Usefuldataset['Aggregate rating'], bins=20, color='cyan')
plt.title('Distribution of Aggregate Ratings')
plt.xlabel('Aggregate Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
Usefuldataset = Usefuldataset.sort_values(by=['Aggregate rating'],ascending=False)
Usefuldataset
```

REMOVING THE RESTAURANTS WITH LOW RATINGS 

```python
Usefuldataset = Usefuldataset[Usefuldataset['Aggregate rating']>3.7]
Usefuldataset
```

```python
Usefuldataset1 = Usefuldataset.copy()
Usefuldataset1['Cuisines'] = Usefuldataset1['Cuisines'].str.split(', ')
Usefuldataset1
```

SHOWING THE MOST COMMON COUSINE USED IN RESTAURANTS

```python
restoXcuisines = pd.crosstab(Usefuldataset1['Restaurant ID'], Usefuldataset1['Cuisines'])
restoXcuisines
```

```python
merged_df = pd.merge(Usefuldataset, restoXcuisines, on='Restaurant ID', how='inner')
merged_df.drop('Cuisines', axis=1, inplace=True)
merged_df
```

CONTENT BASED FILTERING

```python
from sklearn.metrics.pairwise import cosine_similarity

restaurant_features = merged_df.drop(['Restaurant ID', 'Restaurant Name', 'Aggregate rating', 'Votes'], axis=1)

cosine_sim = cosine_similarity(restaurant_features, restaurant_features)

def get_recommendations(restaurant_id, cosine_sim=cosine_sim):
    idx = dataset[dataset['Restaurant ID'] == restaurant_id].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    restaurant_indices = [i[0] for i in sim_scores]
    return dataset[['Restaurant Name', 'Aggregate rating', 'Cuisines']].iloc[restaurant_indices]
```

TESTING THE RECOMMENDATION SYSTEM

```python
recommended_restaurants = get_recommendations(6317637)
print(recommended_restaurants)
```
